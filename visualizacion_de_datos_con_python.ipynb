### Desafio

Trabajas como Analista de Datos en una empresa minorista y te asignaron la tarea de crear una figura con subgráficos que muestre la variación en el número de ventas en cuatro tiendas diferentes a lo largo de un año. La gerencia de la empresa necesita visualizar claramente las tendencias de ventas en cada tienda, para que puedan tomar decisiones estratégicas sobre inventarios y acciones de marketing. Para ello, debes crear cuatro subgráficos dispuestos en dos filas y dos columnas, donde cada subgráfico represente una tienda diferente. En este desafío, cada subgráfico debe presentar un gráfico de líneas que muestre la variación del número de ventas a lo largo de los meses del año.

¡Ahora es el momento de mostrar tus habilidades en análisis de datos y visualización! Para crear el DataFrame con el número de ventas de las tiendas y crear la figura, utiliza la siguiente información:

```python
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
    'Feb': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'May': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Sep': [240, 160, 290, 130],
    'Oct': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dec': [300, 350, 400, 250]}
```

```python
df_ventas = pd.DataFrame(ventas_2022,index=tiendas)
df_ventas
```

```python
fig, grafico_ventas = plt.subplots(2, 2, figsize=(14, 8))
plt.subplots_adjust(wspace=0.3, hspace=0.4)
fig.suptitle('Ventas del 2022 por mes en las tiendas A, B, C y D')

grafico_ventas[0, 0].plot(df_ventas.loc['A'])
grafico_ventas[0, 0].set_title('Ventas en la tienda A')
grafico_ventas[0, 1].plot(df_ventas.loc['B'])
grafico_ventas[0, 1].set_title('Ventas en la tienda B')
grafico_ventas[1, 0].plot(df_ventas.loc['C'])
grafico_ventas[1, 0].set_title('Ventas en la tienda C')
grafico_ventas[1, 1].plot(df_ventas.loc['D'])
grafico_ventas[1, 1].set_title('Ventas en la tienda D')

for ax in grafico_ventas.flat:
    ax.set_xlabel('Mes')
    ax.set_ylabel('Número de ventas')

plt.show()
```

Algunas ideas de personalización para el grafico del desafio anterior que puedes explorar son:

* Cambiar la posición de los títulos de los subgráficos a la izquierda.
* Aumentar el tamaño de la fuente del título general de la figura para resaltarlo.
* Aumentar el tamaño de los títulos y etiquetas de los ejes de los subgráficos.

```python
# Primero creamos la figura
fig, grafico_ventas = plt.subplots(2, 2, figsize=(14, 8))
# Ajustarmos los espacios entre los graficos
plt.subplots_adjust(wspace=0.4, hspace=0.4)
# Agregar un título general
fig.suptitle('Ventas del 2022 por mes en las tiendas A, B, C y D', fontsize=20)

# Lista de colores
colores = ['purple', 'darkgreen', 'brown', 'red']

for i, ax in enumerate(grafico_ventas.flat):
    ax.plot(df_ventas.loc[df_ventas.index[i]], color=colores[i], lw=3)
    ax.set_title(f'Ventas en la tienda {df_ventas.index[i]}', loc='left', fontsize=16)
    ax.set_xlabel('Mes', fontsize=14)
    ax.set_ylabel('Número de ventas', fontsize=14)
    ax.tick_params(labelsize=12)
    ax.grid(color='lightgrey', linestyle='--')

fig.savefig(ruta + 'Ventas_del_2022_por_mes.png',
            transparent=False,dpi=300,bbox_inches='tight')
```

```python
top_10 = df.sort_values('Total',ascending=False).head(10)
top_10
```

```python
ax = sns.barplot(data=top_10,x='Total',y=top_10.index,orient='h')
ax.set(title='Top 10 países con mayor migración hacia Canadá\n1980 - 2013',
       xlabel='Número de Inmigrantes',
       ylabel='Año')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))

ax = sns.barplot(data=top_10,x='Total',y=top_10.index,orient='h')
ax.set_title('Top 10 países con mayor migración hacia Canadá\n1980 - 2013',
             loc='left',fontsize=16)
ax.set_xlabel('Número de Inmigrantes',fontsize=14)
ax.set_ylabel('')
plt.show()
```

```python
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set_frame_on(False)
  ax.xaxis.set_visible(False)
  ax.tick_params(axis='y',which='both',size=0)
  ax = sns.barplot(data=top_10,x='Total',y= top_10.index,hue=top_10.index,orient='h',
                   palette=palette, legend=False)
  ax.set_title('Top 10 países con mayor migración hacia Canadá\n1980 - 2013',
              loc='left',fontsize=16)
  ax.set_xlabel('Número de Inmigrantes',fontsize=14)
  ax.set_ylabel('')
  for i,j in enumerate(top_10['Total']):
    ax.text(j+20,i,str(j),color='black',fontsize=9,ha='left',va='center')
  plt.show()
```

```python
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set(xticklabels=[])
  ax = sns.barplot(data=top_10,x='Total',y= top_10.index,hue=top_10.index,orient='h',
                   palette=palette, legend=False)
  ax.set_title('Top 10 países con mayor migración hacia Canadá\n1980 - 2013',
              loc='left',fontsize=16)
  ax.set_xlabel('Número de Inmigrantes',fontsize=14)
  ax.set_ylabel('')
  sns.despine()
  for i,j in enumerate(top_10['Total']):
    ax.text(j+20,i,str(j),color='black',fontsize=9,ha='left',va='center')
  plt.show()
```
